```python
# bar graph
wine_data = pd.read_csv(APP_ROOTD + 'winemag-data-130k-v2.csv')
wine_data.head()

flg, ax = plt.subplots()
data = wine_data['points'].value_counts()
points = data.index
frequency = data.values

ax.bar(points, frequency)
ax.set_title("Wine Data")
ax.set_xlabel("points")
ax.set_ylabel("Frequency")
```

```python
data = {
    'Country' : ['USA', 'Canada', 'Germany', 'UK', 'France'],
    'GDP_Per_Capita' : [45000, 42000, 52000, 49000, 47000],
    'Income_Per_capital' : [4000, 5000, 7000, 55000, 60000]
}
df = pd.DataFrame(data)
df.plot(x = 'Country', y = ['GDP_Per_Capita', 'Income_Per_capital'], kind = 'bar')
```

```python
totalDeath = [113855, 37332, 3971, 7473, 33871]
totalRecovery = [773400, 325412, 24999, 129812, 34720]
activeCases = [117800, 34987, 239871, 129869, 34987]
country = ['USA', 'Canada', 'Germany', 'UK', 'France']

x = np.arange(len(totalDeath))

plt.bar(x, totalDeath, color = 'black', width = 0.20)
plt.bar(x + 0.21, totalRecovery, color = 'green', width = 0.20)
plt.bar(x + 0.42, totalDeath, color = 'red', width = 0.20)

plt.legend(['Death', 'Recovery', 'Active'])

plt.xticks([i + 0.20 for i in range(5)], country)

plt.title('Bar plot representating total deaths, recovery & active cases')

plt.xlabel('Countries')
plt.ylabel('Cases')

plt.savefig('4BarPlot.png')
plt.show()
```

```python
Data = {
    'Tasks' : [800, 200, 600],
    'Task-Type' : ['Tasks Pending', 'Task Ongoing', 'Task Completed']
}

df = pd.DataFrame(Data)
df.set_index('Task-Type', inplace = True)

df.plot.pie(y = 'Tasks', figsize=(10, 10), autopct = '%1.1f%%', startangle = 90)
```

```python
labels = ['Civil', 'Electrical', 'Mechanical', 'Chemical']
sizes = [15, 60, 50, 40]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
ax.axis('equal')
ax.set_title('Engineering Diciplines')
plt.show()
```

```python
explode = (0.1, 0.1, 0.1, 0.3)
fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=False, startangle=90)
ax.axis('equal')
ax.set_title('Engineering Diciplines')
plt.show()                                   
```
